```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate blood sugar levels (in mg/dL) for 200 patients
# Let's assume two groups: normal (around 100) and diabetic (around 160)
np.random.seed(42)
normal = np.random.normal(100, 15, 120)
diabetic = np.random.normal(160, 20, 80)
blood_sugar = np.concatenate([normal, diabetic])

# Create a KDE (density) plot
sns.kdeplot(blood_sugar, fill=True, color='teal', alpha=0.6)
plt.title("Density Plot of Blood Sugar Levels (200 Patients)")
plt.xlabel("Blood Sugar Level (mg/dL)")
plt.ylabel("Density")
plt.show()
```
